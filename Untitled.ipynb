```python
import pandas as pd
import numpy as np
```

```python
SP = pd.read_csv('data/^GSPC.csv').dropna()
SP = SP[['Date', 'Close']]#[-5000:]
SP.columns = ['date_added', 'gspc']
SP.head()
```

```python
TR = pd.read_csv('data/^TNX.csv').dropna()
TR = TR[['Date', 'Close']]
TR.columns = ['date_added', 'tnx']
TR.head()
```

```python
VIX = pd.read_csv('data/^VIX.csv').dropna()
VIX = VIX[['Date', 'Close']]
VIX.columns = ['date_added', 'vix']
VIX.head()
```

```python
NS = pd.read_csv('data/^IXIC.csv').dropna()
NS = NS[['Date', 'Close']]
NS.columns = ['date_added', 'ixic']
NS.head()
```

```python
RU = pd.read_csv('data/^RUT.csv').dropna()
RU = RU[['Date', 'Close']]
RU.columns = ['date_added', 'rut']
RU.head()
```

```python
WT = pd.read_csv('data/CL=F.csv').dropna()
WT = WT[['Date', 'Close']]
WT.columns = ['date_added', 'wti']
WT.head()
```

```python
GO = pd.read_csv('data/GC=F.csv').dropna()
GO = GO[['Date', 'Close']]
GO.columns = ['date_added', 'gold']
GO.head()
```

```python
# x = pd.concat([SP, VIX], axis=1)
x = pd.merge(SP, VIX, how='left', on='date_added')
x = pd.merge(x, TR, how='left', on='date_added')
x = pd.merge(x, NS, how='left', on='date_added')
x = pd.merge(x, RU, how='left', on='date_added')
x = pd.merge(x, WT, how='left', on='date_added')
x = pd.merge(x, GO, how='left', on='date_added')
# x.Close = x.Close.fillna(0.855)
# x = x[:-1]
x
```

```python
x.tnx = x.tnx.fillna(method='ffill')
x.rut = x.rut.fillna(method='ffill')
x.wti = x.wti.fillna(method='ffill')
x.gold = x.gold.fillna(method='ffill')
# x.tnx = x.tnx.fillna(method='backfill')
# # x = x[8600:]
x.date_added = pd.to_datetime(x.date_added)
# x['date_added'] = x['date_added'].dt.date
# x = x.dropna()
x.info()
x.head(50)
```

```python
# x.to_csv('mybl_ticker.csv', header=False, index=False) #для косольного ввода
x.to_csv('mybl_ticker.csv', index=False)
```
