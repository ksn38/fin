```python
import pandas as pd
import numpy as np
```

```python
SP = pd.read_csv('data/^GSPC.csv').dropna()
SP = SP[['Date', 'Close']]#[8600:]
SP.head()
```

```python
TR = pd.read_csv('data/^TNX.csv').dropna()
TR = TR[['Date', 'Close']]#[8600:]
TR.head()
```

```python
VIX = pd.read_csv('data/^VIX.csv').dropna()
VIX = VIX[['Date', 'Close']]#[8600:]
VIX.head()
```

```python
# x = pd.concat([SP, VIX], axis=1)
x = pd.merge(SP, VIX, how='left', on='Date')
x = pd.merge(x, TR, how='left', on='Date')
# x.Close = x.Close.fillna(0.855)
x = x[:-1]
x
```

```python
x.Close = x.Close.fillna(method='ffill')
# x.Close_y = x.Close_y.fillna(method='backfill')
# x = x[8600:]
x.Date = pd.to_datetime(x.Date)
x.info()
x.head(50)
```

```python
x.columns = ['date_added', 'GSPC', 'VIX', 'TNX']
x.to_csv('mybl_ticker.csv', index=False)
```
