```python
import pandas as pd
import numpy as np
from os import listdir
from os.path import isfile, join
```

```python
mypath = 'C:\\Users\\DFNO\\fin\\hh\\'
files = [i for i in listdir(mypath) if isfile(join(mypath, i))]
print(files)
```

```python
for i in files:
    print(i)
    df = pd.read_csv(mypath + i)
    df.columns = ['tag', 'val']
    print(df)
    df.to_csv(i, index=False, sep=';')
```
